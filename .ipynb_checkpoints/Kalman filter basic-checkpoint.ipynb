```python
import random
import numpy as np
import matplotlib.pyplot as plt

import turtle
import time
```

```python
class KalmanFilter:
    def __init__(self, sigma_x):
        self.P = np.array([[sigma_x*sigma_x, 1000.0*sigma_x], [1000.0*sigma_x, 1000000.0]])
        self.Q = np.ones((2,2)) * 0.1
        self.R = np.ones((1,1)) * (sigma_x*sigma_x)
        self.is_init = False
        
    def predict(self, dt):
        A = np.array([[1.0, dt],[0.0, 1.0]])
        self.x = A * self.x
        A_t = np.transpose(A)
        self.P = A * self.P * A_t + self.Q
    
    def update(self, measurement):
        H = np.array([[1.0, 0.0]])
        H_t = np.transpose(H)
        
        S = H * self.P * H_t + self.R
        S_i = np.linalg.inv(S)
        
        K = self.P * H.T * S_i
        
        z = np.array([[measurement]])
        
        self.x = self.x + K * (z - H * self.x)
        I = np.identity(2)
        self.P = (I - K*H) * self.P
        
    def process(self, measurement, dt):
        if not self.is_init:
            self.x = np.array([[measurement],[0.0]])
            self.is_init = True
            return
        
        self.predict(dt)
        self.update(measurement)

    def state(self):
        return self.x

    def covar(self):
        return self.P
```

```python
sigma = 1.0
v = 1.0
dt = 1.0

x_actual = 10.0
v_actual = v

kf = KalmanFilter(sigma)
no_iter = 30

measured = []
actual = []
position = []
vs = []
covars = []
t = range(no_iter)


for i in t:
    x_measured = random.gauss(x_actual, sigma)
    kf.process(x_measured, dt)

    measured.append(x_measured - x_actual)
    actual.append(x_actual)
    position.append(kf.state()[0][0] - x_actual)
    vs.append(kf.state()[1] - v_actual)
    covars.append(kf.covar()[0][0])

    x_actual += v*dt
```

```python
plt.figure(figsize=(15, 15))
plt.plot(t, measured, 'ro', t, position, 'g', t, covars, 'b')
plt.show()
```
